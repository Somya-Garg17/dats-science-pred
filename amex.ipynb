```python
import torch
import torch.nn as nn
from torch.utils.data import DataLoader, TensorDataset
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import numpy as np

# --- Step 1: Train/Validation Split ---
X_tr, X_val, y_tr, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42)

# --- Step 2: Fill NaNs ---
X_tr = X_tr.fillna(-1)
X_val = X_val.fillna(-1)

# --- Step 3: Convert bool columns to int32 ---
bool_cols = X_tr.select_dtypes(include=['bool']).columns
X_tr[bool_cols] = X_tr[bool_cols].astype('int32')
X_val[bool_cols] = X_val[bool_cols].astype('int32')

# --- Step 4: Convert all columns to float32 ---
X_tr = X_tr.astype('float32')
X_val = X_val.astype('float32')

# --- Step 5: Scale Features ---
scaler = StandardScaler()
X_tr_scaled = scaler.fit_transform(X_tr)
X_val_scaled = scaler.transform(X_val)

# --- Step 6: Convert to PyTorch tensors ---
X_tr_tensor = torch.tensor(X_tr_scaled, dtype=torch.float32)
X_val_tensor = torch.tensor(X_val_scaled, dtype=torch.float32)
y_tr_tensor = torch.tensor(y_tr.values, dtype=torch.float32).view(-1, 1)
y_val_tensor = torch.tensor(y_val.values, dtype=torch.float32).view(-1, 1)


# --- Step 7: DataLoader ---
train_loader = DataLoader(TensorDataset(X_tr_tensor, y_tr_tensor), batch_size=1024, shuffle=True)
val_loader = DataLoader(TensorDataset(X_val_tensor, y_val_tensor), batch_size=1024, shuffle=False)
```

```python
# --- Neural Network ---
class SimpleNN(nn.Module):
    def __init__(self, input_dim):
        super().__init__()
        self.model = nn.Sequential(
            nn.Linear(input_dim, 256),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(256, 128),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(128, 1),
            nn.Sigmoid()  # Use for binary classification
        )

    def forward(self, x):
        return self.model(x)

model = SimpleNN(input_dim=X_tr_tensor.shape[1])
```

```python
criterion = nn.BCELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=5e-3)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = model.to(device)

# Move full-val tensors to device
X_tr_tensor = X_tr_tensor.to(device)
X_val_tensor = X_val_tensor.to(device)
y_tr_tensor = y_tr_tensor.to(device)
y_val_tensor = y_val_tensor.to(device)
train_loader = DataLoader(TensorDataset(X_tr_tensor, y_tr_tensor), batch_size=1024, shuffle=True)
val_loader = DataLoader(TensorDataset(X_val_tensor, y_val_tensor), batch_size=1024, shuffle=False)  
```

```python
import torch
print(torch.__version__)
print(torch.version.cuda)
print(torch.backends.cudnn.version())
print(torch.cuda.is_available())
```

```python
# --- Training Loop ---
for epoch in range(10):
    model.train()
    epoch_loss = 0
    correct = 0
    total = 0

    for xb, yb in train_loader:
        pred = model(xb)
        loss = criterion(pred, yb)
        loss.backward()
        optimizer.step()
        optimizer.zero_grad()

        epoch_loss += loss.item() * xb.size(0)
        predicted = (pred > 0.5).float()
        correct += (predicted == yb).sum().item()
        total += yb.size(0)

    train_loss = epoch_loss / total
    train_acc = correct / total

    model.eval()
    with torch.no_grad():
        val_pred = model(X_val_tensor)
        val_loss = criterion(val_pred, y_val_tensor).item()
        val_acc = ((val_pred > 0.5).float() == y_val_tensor).float().mean().item()

    print(f"Epoch {epoch+1}: "
          f"train_loss = {train_loss:.4f}, train_acc = {train_acc:.4f}, "
          f"val_loss = {val_loss:.4f}, val_acc = {val_acc:.4f}")

```
